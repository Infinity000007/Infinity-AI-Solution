```python
## 100 working code


from fastapi import FastAPI, Request, HTTPException
from fastapi.responses import HTMLResponse
from langchain.vectorstores import Pinecone
import pinecone
from langchain.prompts import PromptTemplate
from langchain.llms import CTransformers
from langchain.chains import RetrievalQA
from dotenv import load_dotenv
from langchain.embeddings import HuggingFaceEmbeddings
import os
from pyngrok import ngrok
import uvicorn
import nest_asyncio

# Apply nest_asyncio patch for compatibility with Jupyter notebooks
nest_asyncio.apply()

# Initialize FastAPI app
app = FastAPI()

# Load environment variables
load_dotenv()

# Access environment variables or use directly
PINECONE_API_KEY = os.getenv("PINECONE_API_KEY", "64bf00ca-a93e-495e-94a9-3dd8f453c960")
PINECONE_ENVIRONMENT = os.getenv("PINECONE_ENVIRONMENT", "us-east-1")
NGROK_AUTH_TOKEN = os.getenv("NGROK_AUTH_TOKEN", "2lh9jbvJmQSk3l7m9JRAq0CuBGu_7sUomed697sTzZEUqYWwf")

# Initialize embeddings and Pinecone client
# Initialize embeddings and Pinecone client
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
pinecone_client = pinecone.Pinecone(api_key=PINECONE_API_KEY, environment=PINECONE_ENVIRONMENT)  # Use pinecone.Pinecone for initialization
index_name = "aibot"
index = pinecone_client.Index(index_name)
docsearch = Pinecone(index, embeddings.embed_query, "text")

# Prompt Template
PROMPT = PromptTemplate(
    template="""Answer the question based on the provided context.

    Context: {context}

    Question: {question}
    Answer:""",
    input_variables=["context", "question"]
)

# Initialize LLM with LLaMA model
llm = CTransformers(
    model="/content/llama-2-7b-chat.ggmlv3.q4_0.bin",
    model_type="llama",
    config={'max_new_tokens': 256, 'temperature': 0.8}  # Reduced token count for faster inference
)

# Setup RetrievalQA chain
qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=docsearch.as_retriever(search_kwargs={'k': 1}),  # Retrieve only the top 1 document
    return_source_documents=False,
    chain_type_kwargs={"prompt": PROMPT}
)

# Define FastAPI routes
@app.get("/", response_class=HTMLResponse)
async def index():
    try:
        with open("/content/new1.html", "r") as file:
            html_content = file.read()
        return HTMLResponse(content=html_content)
    except Exception as e:
        print("Error serving HTML:", e)
        raise HTTPException(status_code=500, detail="Internal Server Error")

@app.post("/get")
async def chat(request: Request):
    try:
        form_data = await request.form()
        msg = form_data.get("msg")
        if not msg:
            raise HTTPException(status_code=400, detail="Message not found in form data")

        print("User Query:", msg)

        # Retrieve documents
        docs = docsearch.similarity_search(msg)
        context = docs[0].page_content if docs else ""

        # Construct the prompt with context and question
        prompt = PROMPT.format(context=context, question=msg)

        # Generate the response
        result = llm(prompt)
        print("Response:", result)

        # Return the response to the frontend
        return {"response": result}
    except Exception as e:
        print("Error:", e)
        raise HTTPException(status_code=500, detail="Internal Server Error")

# Main entry point for running the server
if __name__ == '__main__':
    ngrok.set_auth_token(NGROK_AUTH_TOKEN)
    public_url = ngrok.connect(5000)
    print("Ngrok URL:", public_url)

    uvicorn.run(app, host="0.0.0.0", port=5000, log_level="debug")
```
